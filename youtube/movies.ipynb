```python
likes_percen = youtube['likes'].value_counts(normalize=True, ascending=False) * 100
dislikes_percen = youtube['dislikes'].value_counts(normalize=True, ascending=False) * 100
```

```python
total_likes = youtube['likes'].sum()
total_dislikes = youtube['dislikes'].sum()
total_likes_dislikes = total_likes + total_dislikes
total_likes_dislikes
```

```python
likes_percentage = (total_likes / total_likes_dislikes) * 100
dislikes_percentage = (total_dislikes / total_likes) * 100

data = [likes_percentage, dislikes_percentage]
labels = ['Likes', 'Dislikes']
colors = ['green', 'red']
explode = (0.1, 0)

# plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.figure(figsize=(10, 10))
plt.pie(data, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140, colors=colors, explode=explode, 
        textprops={'fontsize': 14}, labeldistance=1.1, pctdistance=0.6, wedgeprops={'edgecolor': 'black'}, 
        rotatelabels=True, radius=1.5, counterclock=False, center=(0, 0), frame=True)
plt.axis('equal')
plt.title('Likes and dislikes percentage')
plt.show()
```

```python
def views_category(row):
        views = row['views']
        if views < 10_000:
                return 'Low'
        if 10_000 < views < 100_000:
                return 'medium'
        if 100_000 < views < 1_000_000:
                return 'High'
        if views > 1_000_000:
                return 'Very High'
        
youtube.insert(8, 'View category', youtube.apply(views_category, axis=1))
```
