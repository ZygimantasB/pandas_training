## Data Cleaning

```python
study.nunique()
# study.dtypes
```

```python
study['Size'] = study['Size'].fillna('0').str.lower()
```

```python
def split_number_unit(spliter):
    number = ''.join(re.findall(r'\d+\.?\d*', spliter))
    unit = ''.join(re.findall(r'[a-zA-Z]+', spliter))
    return pd.Series([number, unit])
```

```python
study[['Size_number', 'Size Unit']] = study['Size'].apply(split_number_unit)
```

```python
study['Size_number'] = study['Size_number'].str.replace('', '0').astype(float)

for number_col in ['Store', 'Quantity']:
    study[number_col] = study[number_col].astype(int)

for col in ['Store', 'Brand', 'VendorNumber', 'PONumber']:
    study[col] = pd.to_numeric(study[col], downcast='unsigned')

for col_category in ['Store', 'VendorName', 'Size Unit']:
  study[col_category] = study[col_category].astype('category')
```

```python
study.drop('Size', axis=1, inplace=True)

new_columns =  ['InventoryId', 'Store', 'Brand', 'Description', 'Size_number', 'Size Unit', 'VendorNumber','VendorName', 'PONumber', 
                'PODate', 'ReceivingDate', 'InvoiceDate', 'PayDate', 'PurchasePrice', 'Quantity', 'Dollars', 
                'Classification']

study = study[new_columns]
```

```python
rename_column_dict = {
                      'Size_number': 'SizeNumber', 
                      'PONumber' : 'PurchaseOrderNumber',
                      'PODate' : 'PurchaseOrderDate',
                      'Size Unit': 'SizeUnit'
                      }

study = study.rename(columns=rename_column_dict)
```

```python
study['SizeUnit'] = study['SizeUnit'].cat.rename_categories({'mlpk': 'ml'})
```

```python
# Define a dictionary with conversion rates to liters
conversion_dict = {'ml': 0.001, 'l': 1, 'liter': 1, 'oz': 0.0295735, 'gal': 3.78541, '': 0}

# Replace each unit in 'SizeUnit' with its equivalent in liters
study['SizeUnit'] = study['SizeUnit'].map(conversion_dict)

# Multiply 'SizeNumber' by 'SizeUnit' to get all sizes in liters
study['SizeNumber'] = study['SizeNumber'] * study['SizeUnit']
```

```python
study.drop('SizeUnit', axis=1, inplace=True)
```
