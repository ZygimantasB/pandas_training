```python
import pandas as pd
from string import punctuation
```

```python
df_path = r'F:\DataSpell\Pandas_training\Youtube_videos_training\Data Cleaning\data\Customer Call List.xlsx'
```

```python
df = pd.read_excel(df_path)
df
```

```python
df = df.drop_duplicates()
```

```python
# df.drop('Not_Useful_Column', axis=1, inplace=True)
```

```python
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df['Last_Name'] = df['Last_Name'].str.strip(punctuation)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '0', regex=True)

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

df['Phone_Number'] = df['Phone_Number'].apply(lambda 
                      x: x[0:3] + '-' + 
                          x[3:6] + '-' + 
                          x[3:6] + '-' +
                          x[6:10])

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')

df['Phone_Number']  = df['Phone_Number'].str.replace('Na--', '')
```

```python
df
```

```python
df[['Street Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer'] = (df['Paying Customer'].str.replace('Yes', 'Y').str.replace('No', 'N'))
df
```

```python
# df.fillna('')
df = df.replace('N/a', '')
df = df.replace('NaN', '')
```

```python
df
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)
    
df
```

```python
df['Do_Not_Contact'] = df.replace('', 'N')
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == 'Y':
        df.drop(x, inplace=True)

df
```
