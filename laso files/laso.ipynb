```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import laspy

# Load the LAS file
file_path = r"C:\Users\Zygim\Downloads\2743_1234.las"
las = laspy.read(file_path)

# Extract coordinates
x = las.points['X']
y = las.points['Y']
z = las.points['Z']

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=1)  # s is the size of the point

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show plot
plt.show()
```
