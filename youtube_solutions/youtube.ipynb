```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from vega_datasets import data

sp = data.sp500()
sp
```

```python
sp.set_index(['date'], inplace=True)
sp
```

```python
sp[['price']].resample('H').ffill()
```

```python
avg_month = sp[['price']].resample('M').mean()
avg_month
```

```python
plt.figure(figsize=(15, 8))
sns.lineplot(data=avg_month, x='date', y='price')
```

```python
quarter_low = sp[['price']].resample('Q').min()
quarter_low
```

```python
plt.figure(figsize=(15, 8))
sns.lineplot(data=quarter_low, x='date', y='price')
```

```python
quarter_low = sp[['price']].resample('D').min()
quarter_low
```
